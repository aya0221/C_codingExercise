```python
import pandas as pd

df = pd.read_csv('https://techacademy.s3.amazonaws.com/bootcamp/ai/4/files/kadai/4/baseball_salary.csv')
type(df)
df 
```

```python
list(df.columns)
```

```python
df["一安打当たりの年俸"] = df["推定年俸"] / df["打席数"]
df.head()
```

```python
specificColumns = df[["球団", "一安打当たりの年俸",]]

specificColumns
```

```python
groupedTeam = specificColumns.groupby("球団")
temp = groupedTeam.sum()
sortedTeam = temp.sort_values('一安打当たりの年俸', ascending=False)
print(sortedTeam)
```
